### Change data types

```python
# Converting price column from string to float format

# Remove the dollar sign and commas from the price column
listings.loc[:, 'price'] = listings['price'].replace({'\$': '', ',': ''}, regex=True)

# Convert the price column to float
listings.loc[:, 'price'] = listings['price'].astype(float)
```

```python
# Converting date columns to datetime format 

# Convert 'host_since' to datetime 
listings.loc[:, 'host_since'] = pd.to_datetime(listings['host_since'])

# Convert 'first_review' to datetime
listings.loc[:, 'first_review'] = pd.to_datetime(listings['first_review'])

# Convert 'last_review' to datetime 
listings.loc[:, 'last_review'] = pd.to_datetime(listings['last_review'])
```

```python
# Converting boolean columns to boolean format 

# Define a mapping function that converts 't', 'f' to boolean and handles 'nan'
def convert_to_boolean(value):
    if value == 't':
        return True
    elif value == 'f':
        return False
    else:
        return None  # Handle 'nan' and other unexpected values

# Apply the mapping function to each column using .loc to ensure direct modification
listings.loc[:, 'host_is_superhost'] = listings['host_is_superhost'].apply(convert_to_boolean).astype('boolean')
listings.loc[:, 'host_identity_verified'] = listings['host_identity_verified'].apply(convert_to_boolean).astype('boolean')
listings.loc[:, 'has_availability'] = listings['has_availability'].apply(convert_to_boolean).astype('boolean')
listings.loc[:, 'instant_bookable'] = listings['instant_bookable'].apply(convert_to_boolean).astype('boolean')
```

```python
# Convert 'host_response_rate' and 'host_acceptance_rate' from percentage strings to float
listings.loc[:, 'host_response_rate'] = listings['host_response_rate'].str.rstrip('%').astype(float) / 100
listings.loc[:, 'host_acceptance_rate'] = listings['host_acceptance_rate'].str.rstrip('%').astype(float) / 100
```
